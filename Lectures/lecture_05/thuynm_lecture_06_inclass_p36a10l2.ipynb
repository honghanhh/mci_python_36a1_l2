```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import RobustScaler
```

```python
X,y= df.drop(columns=['target','target_binary']),df['target_binary']
```

```python
scaler = RobustScaler() 
X_scaled = scaler.fit_transform(np.log1p(X))
X_scaled
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
x_train, x_test, y_train, y_test=train_test_split(X_scaled,y,test_size=0.2,stratify=y,random_state = 420)
```

```python
forest = RandomForestClassifier()
forest.fit(x_train,y_train)
y_pred = forest.predict(x_test)
```

```python
print(metrics.accuracy_score(y_test, y_pred),metrics.precision_score(y_test,y_pred),metrics.recall_score(y_test,y_pred))
```

```python
from sklearn.model_selection import RepeatedKFold,cross_validate
```

```python
cv = RepeatedKFold(n_splits = 5,n_repeats=6,random_state =420)
scoring = ['accuracy', 'precision', 'recall', 'f1']
forest_scores = cross_validate(forest, X_scaled, y, cv=cv, n_jobs=2, scoring=scoring)
```

```python
sns.scatterplot(data=forest_scores, x='test_accuracy', y='test_f1', s=300, alpha=0.7)
```

```python
sns.scatterplot(data=svc_scores, x='test_accuracy', y='test_f1', s=300, alpha=0.7)
sns.scatterplot(data=forest_scores, x='test_accuracy', y='test_f1', s=300, alpha=0.7)
```
